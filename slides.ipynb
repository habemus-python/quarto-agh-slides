## Part 1: Markdown table alignment

Quarto supports normal Markdown tables.

Alignment is controlled with colons in the separator row:

```markdown
| Label | Value | Unit |
|:------|------:|:----:|
| mass  | 1.23  | kg   |
| time  | 12.5  | s    |
```

Result:

| Label | Value | Unit |
|:------|------:|:----:|
| mass  | 1.23  | kg   |
| time  | 12.5  | s    |

## Part 2: Generate tables from Python

A `pandas.DataFrame` can be displayed directly in Quarto.

This is useful when the table is produced by calculations rather than typed manually.

````python
raw[["sample", "mass_kg", "time_s"]]
````

## Part 3: Number formatting

The tutorial will explain common Python format specifiers:

| Specifier | Meaning | Example |
|:--|:--|:--|
| `.2f` | 2 digits after decimal point | `12.35` |
| `.3g` | 3 significant digits | `0.00432` |
| `.2e` | scientific notation | `1.23e-04` |
| `,.2f` | thousands separator | `1,234.57` |

````python
formatted = raw.copy()

formatted["mass_kg"] = formatted["mass_kg"].map(lambda x: f"{x:.3g}")
formatted["time_s"] = formatted["time_s"].map(lambda x: f"{x:.2f}")
formatted["uncertainty"] = formatted["uncertainty"].map(lambda x: f"{x:.2e}")
formatted["measured_at"] = formatted["measured_at"].dt.strftime("%H:%M")

formatted
````

## Part 4: Significant digits

A key point will be the difference between decimal places and significant digits.

```python
f"{x:.3f}"  # 3 decimal places
f"{x:.3g}"  # 3 significant digits
```

This matters for measurements, where values may have very different magnitudes.

````python
values = [1234.567, 12.34567, 0.1234567, 0.0001234567]

pd.DataFrame({
    "raw value": values,
    "3 decimal places": [f"{x:.3f}" for x in values],
    "3 significant digits": [f"{x:.3g}" for x in values],
})
````

## Part 5: Time values

Time data should usually be converted from full timestamps to a format that matches the purpose of the table.

Examples:

| Format | Meaning |
|:--|:--|
| `%H:%M` | hour and minute |
| `%Y-%m-%d` | date |
| `%Y-%m-%d %H:%M` | date and time |

````python
time_demo = raw[["sample", "measured_at"]].copy()

time_demo["time only"] = time_demo["measured_at"].dt.strftime("%H:%M")
time_demo["date only"] = time_demo["measured_at"].dt.strftime("%Y-%m-%d")
time_demo["date and time"] = time_demo["measured_at"].dt.strftime("%Y-%m-%d %H:%M")

time_demo
````

## Part 6: Polished table with labels

The final example will rename columns so that the table reads like a publication table.

````python
final_table = formatted.rename(columns={
    "sample": "Sample",
    "mass_kg": "Mass / kg",
    "time_s": "Time / s",
    "measured_at": "Measured at",
    "uncertainty": "Uncertainty",
})

final_table
````
